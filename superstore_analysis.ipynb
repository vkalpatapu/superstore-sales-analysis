```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
print('Shape:', df.shape)
print(df.head())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Sales by Category
category_sales = df.groupby('Category')['Sales'].sum().round(2).sort_values(ascending=False)
print(category_sales)
```

```python
# Sales by Category - Bar Chart
category_sales.plot(kind='bar', color='steelblue', figsize=(8,5))
plt.title('Total Sales by Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Profit by Region
region_profit = df.groupby('Region').agg({'Sales':'sum','Profit':'sum'}).round(2)
region_profit['margin_pct'] = (region_profit['Profit'] / region_profit['Sales'] * 100).round(2)
print(region_profit.sort_values('Sales', ascending=False))
```

```python
# Loss-making Sub-Categories
sub_profit = df.groupby('Sub-Category')['Profit'].sum().round(2).sort_values()
print('Loss-making sub-categories:')
print(sub_profit[sub_profit < 0])
```

```python
# Profit by Sub-Category - Bar Chart with color
colors = ['red' if x < 0 else 'green' for x in sub_profit.values]
sub_profit.plot(kind='barh', color=colors, figsize=(10,8))
plt.title('Profit by Sub-Category (Red = Loss)')
plt.xlabel('Total Profit ($)')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
```

```python
# Customer Segment Analysis
segment = df.groupby('Segment').agg(
    total_orders=('Order ID', 'count'),
    total_sales=('Sales', 'sum'),
    avg_order_value=('Sales', 'mean')
).round(2)
print(segment.sort_values('total_sales', ascending=False))
print('\nKey insight: Home Office has highest AOV at $240 despite fewest orders')
```
